```python
import random
for _ in range(10):
  w = random.randint(1,10)
  print(w)

max_weight = 0
max_edge = [0,0]

f = open("graph.txt","r")
v_e = f.readline()
vertex = int(v_e.split()[0])
edges = int(v_e.split()[1])
print(vertex,edges)

matrix =[]
for j in range(vertex):
  row =[]
  for i in range(edges):
    row.append(0)
  matrix.append(row)

for line in f.readlines():
  u = int(line.split()[0])
  v = int(line.split()[1])
  w = random.randint(1,10)
  matrix [u][v] = w
  matrix[v][u] = w

  if w>= max_weight:
    max_weight = w
    max_edge = [u,v]

def MaxDegree(M,u):
  count = 0
  for i in range(len(matrix[u])):
    if matrix[u][i] != 0:
      count += 1
  return count

maxdeg = MaxDegree(matrix,0)

def Neighbor(M,u):
  L = []
  for i in range(len(matrix[u])):
    if matrix[u][i] != 0:
      L.append(i)
  return L
adj = Neighbor(matrix,1)

for row in matrix:
  print(row)



print(max_edge)
print(maxdeg)
print(adj)
```

```python
import random
for _ in range(5):
  w = random.randint(1,10)
  print(w)
f = open ("graph.txt","r")
v_e = f.readline()
veList =v_e.split()
vertex = int(veList[0])
edges = int (veList[1])

print (vertex,edges)

List = {}

for i in range (vertex):
    List[i] =[]

for i in range (edges):
    line = f.readline()
    u = int(line.split()[0])
    v = int(line.split()[1])
    w = random.randint(1,10)
    List[u].append((v,w))
print(List)

```
