```python
#Read From File

f = open("graph.txt","r")
v_e = f.readline()
vertex = int(v_e.split()[0])
edges = int(v_e.split()[1])
print(vertex,edges)


#Create Empty List

matrix =[]
for j in range(vertex):
  row =[]
  for i in range(vertex):
    row.append(0)
  matrix.append(row)
#print(matrix)

# Populate Graph data

for line in f.readlines():
  u = int(line.split()[0])
  v = int(line.split()[1])
  matrix [u][v] = 1

for row in matrix:
  print(row)
```

```python
#Adjecent & Degree
#Read From File

f = open("graph.txt","r")
v_e = f.readline()
vertex = int(v_e.split()[0])
edges = int(v_e.split()[1])
print(vertex,edges)


#Create Empty List

matrix =[]
for j in range(vertex):
  row =[]
  for i in range(vertex):
    row.append(0)
  matrix.append(row)
#print(matrix)

# Populate Graph data

for line in f.readlines():
  u = int(line.split()[0])
  v = int(line.split()[1])
  matrix [u][v] = 1

for row in matrix:
  print(row)

def adjecent(M,u):
  L = []
  for i in range(len(matrix[u])):
    if matrix[u][i] ==1:
      L.append(i)
  return L
adj = adjecent(matrix,0)
print(adj)


def degree(M,u):
  count = 0
  for i in range(len(matrix[u])):
    if matrix[u][i] ==1:
      count += 1
  return count
adj = adjecent(matrix,1)
deg = degree(matrix,1)
print(adj)
print(deg)
```

```python
#Adjecent & Degree

#Directed graph

#Read graph data
f = open("graph3.txt","r")
v_e = f.readline()
veList = v_e.split()
vertex = int(veList[0])
edges = int(veList[1])
print(vertex,edges)

#Creat Empty List

List = {}
for i in range(vertex):
  List[i] = []
print(List)

# Populate the graph

for i in range(edges):
  line = f.readline()
  u = int(line.split()[0])
  v = int(line.split()[1])
  List[u].append(v)

def adjecency(L,u):
  return L[u]

def outDegree(L,u):
  return len(L[u])

def inDegree(L,u):
  c = 0
  for i in range(len(L)):
    for j in range(len(L[i])):
      if L[i][j] == u:
        c += 1
  return c



print(List)
adj = adjecency(List,4)
print(adj)
Odeg = outDegree(List,4)
print(Odeg)
indeg = inDegree(List,4)
print(indeg)
```
