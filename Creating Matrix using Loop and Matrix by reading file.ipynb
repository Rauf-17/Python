```python
import random
vertex = input("Vertex: ")
No_of_edges = input("Edges: ")
f = open("wg.txt","w")
f.write(vertex +" "+No_of_edges)
i = 0
while i<int(No_of_edges):
  u = random.randint(0,int(vertex)-1)
  v = random.randint(0,int(vertex)-1)
  if u!=v:
    f.write("\n")
    f.write(str(u)+" "+str(v))
    i = i+1
f.close()
```
